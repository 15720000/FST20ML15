# 4. Tests whether a data sample has a Gaussian distribution.
a)	Generate the data, form H0 and H1 and write python code to test the Normality        [10M]


```python
from scipy.stats import shapiro

# generate some data
data = [0.25, 0.8, 0.9, 0.26, 0.53]

# form the null hypothesis and alternative hypothesis
# H0: the data is normally distributed
# H1: the data is not normally distributed
alpha = 0.05

# perform the Shapiro-Wilk test
stat, p = shapiro(data)

# print the results
print('Statistics=%.3f, p=%.3f' % (stat, p))
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')
```

# Import Dataset

```python
dataset=pd.read_csv(r"C:\Users\Lenovo\Downloads\CardioGoodFitness.csv")
dataset
```

```python
from sklearn.preprocessing import LabelEncoder                        # Labelling the features.
l1=LabelEncoder()
dataset["Gender"]=l1.fit_transform(dataset["Gender"])
dataset["MaritalStatus"]=l1.fit_transform(dataset["MaritalStatus"])
```

```python
#b.Which gender has better fitness levels?       
#calculate the mean fitness score for each gender
mean_fitness_by_gender = dataset.groupby('Gender')['Fitness'].mean()        #Male denoted by 1 and female denote by 0.
print(mean_fitness_by_gender)
```

# Interpretation:Male has better fitness levels.

```python
#c.Which is the most popular product among all?
#count the number of occurrences of each product
product_counts = dataset['Product'].value_counts()
print(product_counts)
#categorical variables 
plt.figure(figsize=(5,5))
dataset['Product'].value_counts().plot.pie(autopct='%1.1f%%',figsize=(8,8))
plt.title("Pie chart of Product Sales")
plt.show()
```

# Interpretation:TM195 is the most popular product among all.

```python
#d.Who is more fit? Married or singles
#calculate the mean fitness score for each marital status
mean_fitness_by_marital_status = dataset.groupby('MaritalStatus')['Fitness'].mean() # single is denoted by 1 and partnered is denoted by 0.
print(mean_fitness_by_marital_status)
```

# Interpretation: single is more fit than the partnered.

```python
#e.Which is the most favourite product of the customers below the age of 40?
#Here create a subset of the dataframe for customers below the age of 40
dataset_below_40 = dataset[dataset['Age'] < 40]
# count the number of occurrences of each product in the subset
product_counts_below_40 = dataset_below_40['Product'].value_counts()
print(product_counts_below_40)
```

# Interpretation: TM195 is the most favourite product of the customers below the age of 40

# 5.Python Program to Make a Simple Calculator



```python
# function to add two numbers
def add(x, y):
    return x + y

# function to subtract two numbers
def subtract(x, y):
    return x - y

# function to multiply two numbers
def multiply(x, y):
    return x * y

# function to divide two numbers
def divide(x, y):
    return x / y

print("Select operation.")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

# take input from the user
choice = input("Enter choice (1/2/3/4): ")

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

if choice == '1':
    print(num1, "+", num2, "=", add(num1,num2))

elif choice == '2':
    print(num1, "-", num2, "=", subtract(num1,num2))

elif choice == '3':
    print(num1, "*", num2, "=", multiply(num1,num2))

elif choice == '4':
    print(num1, "/", num2, "=", divide(num1,num2))
else:
    print("Invalid input")
```
